```python
pipeline = [
    {
        "$group": {
            "_id": "$duuid",  # Group by field1
            "largest_field2": {"$max": "$EXP"},  # Get the max of field2
            "musername3": {"$last": "$musername"}
        }
    }
]
result = expgains.aggregate(pipeline)
```

```python
%%time
expgains_docs = dict()
expgains_docs_musername = dict()
for item in result:
    # print(f"Field1: {item['_id']}, Largest Field2: {item['largest_field2']}") 
    expgains_docs[item['_id']]=item['largest_field2']
    expgains_docs_musername[item['_id']]=item['musername3']
```

# handles players who have not logged in yet.

```python
%%time
counts=0
docs_to_insert = []
for duuid,val in expgains_docs.items():
    if duuid not in duuid_to_exp:
        musername = expgains_docs_musername[duuid]
        doc = {"duuid":duuid,"musername":musername,"date":datetime.datetime.utcnow(),"EXP":val,"dates":[],"servers":[]}
        docs_to_insert.append( doc )
        counts+=1
```
